```python
# df.iloc[0]

from difflib import SequenceMatcher

def get_overlap_ratio(s1, s2):
    """
    Calculate the ratio of overlapping characters between two strings.
    """
    matcher = SequenceMatcher(None, s1, s2)
    return matcher.find_longest_match(0, len(s1), 0, len(s2)).size

def get_most_overlap_index(predicted_error_sentence, indexed_sentences):
    """
    Get the index of the element in indexed_sentences with the most string overlap with predicted_error_sentence.
    """
    max_overlap = 0
    max_overlap_index = -1
    
    # indexed_sentences

    for i, sentence in enumerate(indexed_sentences):
        try:
            overlap = get_overlap_ratio(predicted_error_sentence, sentence[1][0])
        except:
            print(f"sentence = {sentence}")
            continue
        
        if overlap > max_overlap:
            max_overlap = overlap
            max_overlap_index = i
            
    try: 
        max_overlap_index = int(indexed_sentences[i][0])
    except:
        print(f"indexed_sentences = {indexed_sentences}")

    return max_overlap_index

# # Example usage
# predicted_error_sentence = "This is a test sentence"
# indexed_sentences = ["This is a test", "Another sentence", "Yet another test"]

# most_overlap_index = get_most_overlap_index(predicted_error_sentence, indexed_sentences)
# print("Index with most overlap:", most_overlap_index)
```
